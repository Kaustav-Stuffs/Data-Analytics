# Task2

```python
df = pd.read_csv('dcr.csv')
```

```python
data = df[['FICO_orig_time', 'LTV_orig_time', 'default_time', 'time']]
```

```python
# Remove rows with missing values
data = data.dropna()

# Convert default_time column to a binary indicator (1 for default, 0 for non-default)
data['default_time'] = data['default_time'].astype(bool).astype(int)

# Add a constant column for the intercept
data['intercept'] = 1
```

```python
model = sm.Logit(data['default_time'], data[['intercept', 'FICO_orig_time', 'LTV_orig_time']])
result = model.fit()
print(result.summary())
```

```python
# Predict the default probabilities using the fitted model
data['PD'] = result.predict(data[['intercept', 'FICO_orig_time', 'LTV_orig_time']])

# Group by time and calculate the average PD
average_PD = data.groupby('time')['PD'].mean()
```

```python
plt.plot(average_PD.index, average_PD)
plt.xlabel('Time')
plt.ylabel('Average Probability of Default (PD)')
plt.title('Average Probability of Default Over Time')
plt.show()
```

```python
# Plot 2: Average PD by state over time
average_PD_by_state = data.groupby(['time', 'default_time'])['PD'].mean().unstack()
plt.figure(2)
average_PD_by_state.plot()
plt.xlabel('Time')
plt.ylabel('Average Probability of Default (PD)')
plt.title('Average Probability of Default by State Over Time')

plt.show()
```

# Part 3

```python
# Calculate the loss given default (LGD) for each loan and period
df['LGD'] = df['balance_time'] * (1 - (1 / (1 + 0.4)))  # Assuming 40% house price decline
```

```python
# Filter out non-defaulted loans
defaulted_loans = df[df['default_time'] == 1]
```

```python
# Calculate the hypothetical LGD for defaulted loans assuming no repayment
defaulted_loans['hypothetical_LGD'] = defaulted_loans['balance_time'] * (1 - (1 / (1 + 0.4)))
```

```python
# Plot a histogram for the distribution of LGD
plt.hist(defaulted_loans['hypothetical_LGD'], bins=20)
plt.xlabel('Loss Given Default (LGD)')
plt.ylabel('Frequency')
plt.title('Distribution of LGD')
plt.show()
```

```python
# Analyze the output
# Compute the average LGD for all defaulted loans
average_LGD = defaulted_loans['hypothetical_LGD'].mean()
```

```python
# Compute the average LGD by time period
average_LGD_by_period = defaulted_loans.groupby('time')['hypothetical_LGD'].mean()

print("Average LGD for all defaulted loans: ", average_LGD)
print("Average LGD by time period:\n", average_LGD_by_period)
```

```python
# Select the relevant columns for the regression model
data = df[['LGD', 'FICO_orig_time', 'LTV_orig_time', 'default_time', 'time']]
```

```python
# Remove rows with missing values
data = data.dropna()

# Add a constant column for the intercept
data['intercept'] = 1
```

```python
# Fit the linear regression model
model = sm.OLS(data['LGD'], data[['intercept', 'FICO_orig_time', 'LTV_orig_time']])
result = model.fit()
```

```python
# Print the regression results
print(result.summary())
```

```python
# Predict the LGD using the fitted model
data['predicted_LGD'] = result.predict(data[['intercept', 'FICO_orig_time', 'LTV_orig_time']])
```

```python
# Group by time and calculate the average predicted LGD
average_predicted_LGD = data.groupby('time')['predicted_LGD'].mean()
```

```python
# Plot the average LGD by period
plt.plot(average_predicted_LGD.index, average_predicted_LGD)
plt.xlabel('Time')
plt.ylabel('Average LGD')
plt.title('Average LGD by Period')
plt.show()
```

```python
# Analyze the output

# Compute the average predicted LGD for all mortgage loans and periods
average_predicted_LGD_all = data['predicted_LGD'].mean()

print("Average predicted LGD for all mortgage loans and periods:", average_predicted_LGD_all)
```

```python
# Extract necessary columns from mortgage_data
df = df[['time', 'balance_time']]

# Calculate capital ratio
df['capital_ratio'] = df['balance_time'] / df['balance_time'].max()

# Group data by time and calculate average capital ratio
average_capital_ratio = df.groupby('time')['capital_ratio'].mean().reset_index()

# Plot average capital ratio over time
plt.figure(figsize=(10, 6))
plt.plot(average_capital_ratio['time'], average_capital_ratio['capital_ratio'], marker='o')
plt.xlabel('Time')
plt.ylabel('Average Capital Ratio')
plt.title('Average Capital Ratio Over Time')
plt.grid(True)
plt.show()
```

```python
# Plot average balance over time
plt.figure(figsize=(10, 6))
plt.plot(df['time'], df['balance_time'], marker='o')
plt.xlabel('Time')
plt.ylabel('Average Balance')
plt.title('Average Balance Over Time')
plt.grid(True)
plt.show()
```
